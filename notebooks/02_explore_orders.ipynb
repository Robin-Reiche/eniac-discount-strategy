## Documentation

### Dataset Overview
- **226,909 rows** and **4 columns:** `order_id`, `created_date`, `total_paid`, `state`
- `order_id` has no duplicates
- `state` contains these values: 'Cancelled', 'Completed', 'Pending', 'Shopping Basket', 'Place Order'

### Data Quality Notes
- **5 null values** in `total_paid` - all with status 'Pending'
- `total_paid` is often 0 (~20,000 cases) which indicates a potential data issue:
  - Shopping Basket: 9,858
  - Place Order: 9,419
  - Cancelled: 35
  - Completed: 16
  - Pending: 6

### Summary Statistics (without data cleaning)
- **Average order value:** $569.23 USD
- **Median:** $112.99 USD
- **75th percentile:** $525.98 (three quarters of all orders are below this value)
- **10 orders** exceeded $100,000 USD
