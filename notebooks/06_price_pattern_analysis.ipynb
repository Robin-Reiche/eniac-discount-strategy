```python
import pandas as pd

# Load data using relative paths
brands = pd.read_csv('../data/raw/brands.csv')
orderlines = pd.read_csv('../data/raw/orderlines.csv')
orders = pd.read_csv('../data/raw/orders.csv')
products = pd.read_csv('../data/raw/products.csv')

# Create working copies
brands_df = brands.copy()
orderlines_df = orderlines.copy()
orders_df = orders.copy()
products_df = products.copy()
```

```python
products_df_copy = products_df.copy()
```

```python
merged_df = products_df_copy.merge(orderlines_df, on='sku')
merged_df = merged_df.loc[:,['sku', 'name', 'desc', 'price', 'promo_price', 'unit_price']]
merged_df.info()
merged_df.head()
merged_df = merged_df.drop_duplicates()
merged_df.duplicated().sum()
```

```python
total_len = len(merged_df)
dot_pattern = r'\.'
print(f"over 2 dots: {(merged_df['unit_price'].str.count(dot_pattern) > 2).sum() / total_len * 100}")
print(f"over 1 dot: {(merged_df['unit_price'].str.count(dot_pattern) > 1).sum() / total_len * 100}")
print(f"exactly 1 dot: {(merged_df['unit_price'].str.count(dot_pattern) == 1).sum() / total_len * 100}")
print(f"no dots: {(merged_df['unit_price'].str.count(dot_pattern) == 0).sum() / total_len * 100}")
```

```python
mask_1_dot = merged_df['unit_price'].str.count(r'\.') == 1
merged_df.loc[mask_1_dot,:].sample(20)
```

```python
mask_2_dots = merged_df['unit_price'].str.count(r'\.') > 1
merged_df.loc[mask_2_dots,:].sample(20)
```

```python
total_len = len(merged_df)
print(total_len)
print((merged_df['unit_price'].str.extract(r'\.(\d+)$')[0].str.len() == 2).sum())
```

```python
# Fix unit_price format (remove extra dots for thousand separators)
merged_df['unit_fixed'] = merged_df['unit_price'].copy()
mask = merged_df['unit_price'].str.count(r'\.') > 1
merged_df.loc[mask, 'unit_fixed'] = merged_df.loc[mask, 'unit_price'].str.replace(r'\.', '', n=1, regex=True)
merged_df['unit_fixed'] = pd.to_numeric(merged_df['unit_fixed'])
print(merged_df['unit_fixed'].isna().sum())
merged_df.loc[mask, :].sample(10)
```

```python
merged_df['unit_fixed'].nsmallest(10)
negative_unit_fixed_mask = merged_df['unit_fixed'] <= 0
merged_df.loc[negative_unit_fixed_mask, :].sort_values(by='unit_fixed', ascending=True)
# TODO: discuss with team how to handle zero prices
```

```python
negative_unit_fixed_mask = merged_df['unit_fixed'] < 0
merged_df.loc[negative_unit_fixed_mask, ['unit_price', 'unit_fixed']] *= -1
merged_df.isna().sum()
```

```python
merged_df.isna().sum()
isna_mask = merged_df['price'].isna()
merged_df.loc[isna_mask, :]
```

```python
merged_df["price"] = merged_df["price"].fillna(merged_df["unit_fixed"])
merged_df.isna().sum()
```

```python
merged_df['price_unit_fixed_ratio'] = pd.to_numeric(merged_df['price'], errors='coerce') / merged_df['unit_fixed']
merged_df[['price', 'unit_price', 'unit_fixed', 'price_unit_fixed_ratio']].nsmallest(10, 'price_unit_fixed_ratio')
```

```python
merged_df[['price', 'unit_price', 'unit_fixed', 'price_unit_fixed_ratio']].sort_values(by='price_unit_fixed_ratio', ascending=False).tail(100)
merged_df[(merged_df['price_unit_fixed_ratio'] > 0) & (merged_df['unit_fixed'] > 0)][['price', 'unit_price', 'unit_fixed', 'price_unit_fixed_ratio']].sort_values(by='price_unit_fixed_ratio', ascending=False).head(1000)
# TODO: discuss with team how to handle extreme price ratios
```

```python
merged_df['reliable_price_and_unit'] = (merged_df['price_unit_fixed_ratio'] >= 0.4) & (merged_df['price_unit_fixed_ratio'] <= 5)
merged_df['reliable_price_and_unit'].value_counts(normalize=True).to_frame()
```

```python
merged_df['price_dots'] = merged_df['price'].str.count(r'\.')
merged_df['price_dots'].value_counts().to_frame().sort_values(by='count',ascending=False)
```

```python
merged_df.groupby('price_dots')['reliable_price_and_unit'].value_counts().to_frame()
```

```python
merged_df.groupby('price_dots')['reliable_price_and_unit'].value_counts(normalize=True).to_frame() * 100
```

```python
mask = merged_df['price_dots'] == 2
merged_df.loc[mask,:].sample(10)
```

```python
merged_df['price_fixed'] = pd.to_numeric(merged_df['price'], errors='coerce')
mask_2plus_dots = merged_df['price_dots'] > 1
temp_mask = merged_df.loc[mask_2plus_dots, 'price'].str.replace(r'\.', '', n=1, regex=True)
merged_df.loc[mask_2plus_dots, 'price_fixed'] = pd.to_numeric(temp_mask, errors='coerce') / 10
merged_df['price_fixed'] = merged_df['price_fixed'].round(2)
merged_df.sample(10)
```

```python
merged_df['price_unit_fixed_ratio_new'] = merged_df['price_fixed'] / merged_df['unit_fixed']
merged_df['reliable_price_and_unit_new'] = (merged_df['price_unit_fixed_ratio_new'] >= 0.4) & (merged_df['price_unit_fixed_ratio_new'] <= 5)
merged_df.groupby('price_dots')['reliable_price_and_unit_new'].value_counts(normalize=True).to_frame() * 100
```

```python
merged_df[['price', 'price_fixed', 'unit_price', 'unit_fixed', 'price_unit_fixed_ratio', 'price_unit_fixed_ratio_new']].nsmallest(10, 'price_unit_fixed_ratio_new')
```

```python
mask_2plus_dots = merged_df['price_dots'] > 1
merged_df.loc[mask_2plus_dots, :].sample(10)
```

```python
merged_df['promo_price_dots'] = merged_df['promo_price'].str.count(r'\.')
merged_df['promo_price_dots'].value_counts().to_frame().sort_values(by='count',ascending=False)
```

```python
merged_df['promo_price_fixed'] = pd.to_numeric(merged_df['promo_price'], errors='coerce')
mask_2plus_dots = merged_df['promo_price_dots'] > 1
temp_mask = merged_df.loc[mask_2plus_dots, 'promo_price'].str.replace(r'\.', '', n=1, regex=True)
merged_df.loc[mask_2plus_dots, 'promo_price_fixed'] = pd.to_numeric(temp_mask, errors='coerce') / 10
merged_df['promo_price_fixed'] = merged_df['promo_price_fixed'].round(2)
merged_df.sample(10)
```

```python
merged_df['promo_price_to_unit_price_ratio'] = merged_df['promo_price_fixed'] / merged_df['unit_fixed']
merged_df['reliable_promo_price_to_unit_price'] = (merged_df['promo_price_to_unit_price_ratio'] >= 0.4) & (merged_df['promo_price_to_unit_price_ratio'] <= 5)
merged_df.groupby('promo_price_dots')['reliable_promo_price_to_unit_price'].value_counts(normalize=True).to_frame() * 100
```

```python
merged_df.loc[
    merged_df["reliable_promo_price_to_unit_price"] == 0,
    [
        "price",
        "price_fixed",
        "promo_price",
        "promo_price_fixed",
        "unit_price",
        "unit_fixed",
    ],
].sample(10)
```

```python
merged_df['promo_digits_after_dot'] = merged_df['promo_price'].str.extract(r'\.(\d+)$')[0].str.len()
merged_df['promo_digits_after_dot'].value_counts().to_frame().sort_values(by='promo_digits_after_dot', ascending=False)
```

```python
merged_df.loc[
    (merged_df["reliable_promo_price_to_unit_price"] == 0) & (merged_df['promo_digits_after_dot'] == 3),
    [
        "price",
        "price_fixed",
        "promo_price",
        "promo_price_fixed",
        "unit_price",
        "unit_fixed",
        "promo_price_to_unit_price_ratio"
    ],
]
```

```python
digits_mask = (merged_df["promo_price_dots"] == 1) & (
    merged_df["promo_digits_after_dot"] == 3
)
merged_df.loc[
    digits_mask, ["price", "price_fixed", "promo_price", "promo_price_fixed", 'unit_price', 'unit_fixed']].sample(10)
```

```python
merged_df.loc[digits_mask, 'promo_price_fixed'] = merged_df.loc[digits_mask, 'promo_price_fixed'] / 10
```

```python
merged_df['promo_price_to_unit_ratio'] = merged_df['promo_price_fixed'] / merged_df['unit_fixed']
merged_df['reliable_promo_price_to_unit_price'] = (merged_df['promo_price_to_unit_ratio'] >= 0.4) & (merged_df['promo_price_to_unit_ratio'] <= 5)
merged_df.groupby('promo_price_dots')['reliable_promo_price_to_unit_price'].value_counts(normalize=True).to_frame() * 100
```

```python
digits_mask = (merged_df["promo_price_dots"] == 2) & (
    merged_df["promo_digits_after_dot"] == 3) & (merged_df["reliable_promo_price_to_unit_price"] == 0)
merged_df.loc[
    digits_mask, ["price", "price_fixed", "promo_price", "promo_price_fixed", 'unit_price', 'unit_fixed']].sample(10)
```

```python
reliable_df = merged_df[
    (merged_df['reliable_price_and_unit_new'] == True) & 
    (merged_df['reliable_promo_price_to_unit_price'] == True)
].copy()

print(f"Original data: {len(merged_df):,} rows")
print(f"After cleaning (reliable only): {len(reliable_df):,} rows")
print(f"Removed: {len(merged_df) - len(reliable_df):,} rows")
print(f"\nPercentage distribution:")
print(f"  Reliable: {len(reliable_df) / len(merged_df) * 100:.2f}%")
print(f"  Unreliable: {(len(merged_df) - len(reliable_df)) / len(merged_df) * 100:.2f}%")
```
