```python
mask = data_df['price_dots'] == 2
data_df.loc[mask,:].sample(10)
```

```python
data_df['price_fixed'] = pd.to_numeric(data_df['price'], errors='coerce')
mask_2plus_dots = data_df['price_dots'] > 1
temp_mask = data_df.loc[mask_2plus_dots, 'price'].str.replace(r'\.', '', n=1, regex=True)
data_df.loc[mask_2plus_dots, 'price_fixed'] = pd.to_numeric(temp_mask, errors='coerce') / 10
data_df['price_fixed'] = data_df['price_fixed'].round(2)
data_df.sample(10)
```

```python
data_df['price_unit_fixed_ratio_new'] = data_df['price_fixed'] / data_df['unit_fixed']
data_df['reliable_price_and_unit_new'] = (data_df['price_unit_fixed_ratio_new'] >= 0.4) & (data_df['price_unit_fixed_ratio_new'] <= 5)
data_df.groupby('price_dots')['reliable_price_and_unit_new'].value_counts(normalize=True).to_frame() * 100
```

```python
# 1. Examine the problematic 2-dot cases
mask_2dots_unreliable = (data_df['price_dots'] == 2) & (data_df['reliable_price_and_unit_new'] == False)

print(f"Number of unreliable with 2 dots: {mask_2dots_unreliable.sum()}")

problem_cases = data_df.loc[mask_2dots_unreliable, [
    'price', 'price_fixed', 
    'unit_price', 'unit_fixed',
    'price_unit_fixed_ratio_new'
]].copy()

# Check how many are too small vs too large
too_small = (problem_cases['price_unit_fixed_ratio_new'] < 0.4).sum()
too_large = (problem_cases['price_unit_fixed_ratio_new'] > 5).sum()

print(f"\nRatio < 0.4: {too_small}")
print(f"Ratio > 5: {too_large}")

# Show only those that exist
if too_small > 0:
    print("\n=== Ratio TOO SMALL (< 0.4) ===")
    print(problem_cases[problem_cases['price_unit_fixed_ratio_new'] < 0.4].sample(min(10, too_small)))

if too_large > 0:
    print("\n=== Ratio TOO LARGE (> 5) ===")
    print(problem_cases[problem_cases['price_unit_fixed_ratio_new'] > 5].sample(min(10, too_large)))
    
    # Show the distribution
    print(f"\n=== Distribution of too large ratios ===")
    print(problem_cases[problem_cases['price_unit_fixed_ratio_new'] > 5]['price_unit_fixed_ratio_new'].describe())
```

```python
# Analyze the structure of prices
mask_2dots = data_df['price_dots'] == 2
data_df['price_last_digits'] = data_df['price'].str.extract(r'\.(\d+)$')[0].str.len()

print("\n=== Number of digits after last dot for 2-dot prices ===")
print(data_df.loc[mask_2dots, 'price_last_digits'].value_counts())

print("\n=== Examples by digit count ===")
for digits in sorted(data_df.loc[mask_2dots, 'price_last_digits'].unique()):
    mask_digits = mask_2dots & (data_df['price_last_digits'] == digits)
    print(f"\n{digits} digits after last dot:")
    print(data_df.loc[mask_digits, ['price', 'price_fixed', 'unit_price', 'unit_fixed']].sample(min(5, mask_digits.sum())))
```

```python
# Create both variants and choose the one closer to unit_price
data_df['price_fixed'] = pd.to_numeric(data_df['price'], errors='coerce')

mask_2plus_dots = data_df['price_dots'] > 1
temp_str = data_df.loc[mask_2plus_dots, 'price'].str.replace(r'\.', '', n=1, regex=True)
temp_numeric = pd.to_numeric(temp_str, errors='coerce')

# Calculate both options
option_div10 = temp_numeric / 10
option_div100 = temp_numeric / 100

# Calculate absolute difference to unit_fixed
diff_div10 = abs(option_div10 - data_df.loc[mask_2plus_dots, 'unit_fixed'])
diff_div100 = abs(option_div100 - data_df.loc[mask_2plus_dots, 'unit_fixed'])

# Choose the option closer to unit_fixed
data_df.loc[mask_2plus_dots, 'price_fixed'] = option_div10.where(diff_div10 < diff_div100, option_div100)

data_df['price_fixed'] = data_df['price_fixed'].round(2)

# Recalculate ratio
data_df['price_unit_fixed_ratio_new'] = data_df['price_fixed'] / data_df['unit_fixed']
data_df['reliable_price_and_unit_new'] = (data_df['price_unit_fixed_ratio_new'] >= 0.4) & (data_df['price_unit_fixed_ratio_new'] <= 5)

print("\n=== AFTER INTELLIGENT CORRECTION ===")
print(data_df.groupby('price_dots')['reliable_price_and_unit_new'].value_counts(normalize=True).to_frame() * 100)

# Show examples after correction
print("\n=== Examples after correction ===")
print(data_df[data_df['price_dots'] == 2].sample(10)[['price', 'price_fixed', 'unit_price', 'unit_fixed', 'price_unit_fixed_ratio_new', 'reliable_price_and_unit_new']])
```

```python
data_df['promo_price_dots'] = data_df['promo_price'].str.count(r'\.')
data_df['promo_price_dots'].value_counts().to_frame().sort_values(by='count',ascending=False)
```

```python
# Fix promo_price similar to how we fixed price
# First, create promo_price_fixed column (similar to price_fixed logic)

# Step 1: Initialize promo_price_fixed
data_df['promo_price_fixed'] = pd.to_numeric(data_df['promo_price'], errors='coerce')

# Step 2: Fix 2+ dots cases (remove first dot, divide appropriately)
mask_2plus_dots_promo = data_df['promo_price_dots'] > 1
temp_str = data_df.loc[mask_2plus_dots_promo, 'promo_price'].str.replace(r'\.', '', n=1, regex=True)
temp_numeric = pd.to_numeric(temp_str, errors='coerce')

# Use similar logic as price_fixed: divide by 10
data_df.loc[mask_2plus_dots_promo, 'promo_price_fixed'] = temp_numeric / 10
data_df['promo_price_fixed'] = data_df['promo_price_fixed'].round(2)

# Step 3: Calculate initial ratio
data_df['promo_price_to_unit_ratio'] = data_df['promo_price_fixed'] / data_df['unit_fixed']
data_df['reliable_promo_price_to_unit_price'] = (
    (data_df['promo_price_to_unit_ratio'] >= 0.4) & 
    (data_df['promo_price_to_unit_ratio'] <= 5)
)

print("=== Initial promo_price fix results ===")
print(data_df.groupby('promo_price_dots')['reliable_promo_price_to_unit_price'].value_counts(normalize=True).to_frame() * 100)

# Step 4: Additional correction for 1-dot + 3 digits
data_df['promo_digits_after_dot'] = data_df['promo_price'].str.extract(r'\.(\d+)$')[0].str.len()

digits_mask = (data_df["promo_price_dots"] == 1) & (data_df["promo_digits_after_dot"] == 3)
print(f"\n1-dot promo prices with 3 digits: {digits_mask.sum()}")

# Show examples BEFORE correction
print("\n=== BEFORE 3-digit correction ===")
if digits_mask.sum() > 0:
    print(data_df.loc[digits_mask, ['promo_price', 'promo_price_fixed', 'unit_fixed', 'promo_price_to_unit_ratio']].sample(min(10, digits_mask.sum())))

# Correct: divide by 10
data_df.loc[digits_mask, 'promo_price_fixed'] = data_df.loc[digits_mask, 'promo_price_fixed'] / 10
data_df['promo_price_fixed'] = data_df['promo_price_fixed'].round(2)

# Recalculate ratio
data_df['promo_price_to_unit_ratio'] = data_df['promo_price_fixed'] / data_df['unit_fixed']
data_df['reliable_promo_price_to_unit_price'] = (
    (data_df['promo_price_to_unit_ratio'] >= 0.4) & 
    (data_df['promo_price_to_unit_ratio'] <= 5)
)

print("\n=== AFTER 3-digit correction ===")
print(data_df.groupby('promo_price_dots')['reliable_promo_price_to_unit_price'].value_counts(normalize=True).to_frame() * 100)

# Show examples AFTER correction
print("\n=== Sample after correction ===")
if digits_mask.sum() > 0:
    print(data_df.loc[digits_mask, ['promo_price', 'promo_price_fixed', 'unit_fixed', 'promo_price_to_unit_ratio', 'reliable_promo_price_to_unit_price']].sample(min(10, digits_mask.sum())))
```

```python
# Create final reliable DataFrame
reliable_df = data_df[
    (data_df['reliable_price_and_unit_new'] == True) & 
    (data_df['reliable_promo_price_to_unit_price'] == True)
].copy()

print("=" * 60)
print("FINAL DATA CLEANING - SUMMARY")
print("=" * 60)

print(f"\nOriginal data: {len(data_df):,} rows")
print(f"After cleaning (reliable only): {len(reliable_df):,} rows")
print(f"Removed: {len(data_df) - len(reliable_df):,} rows")

print(f"\n{'Percentage distribution:'}")
print(f"  ✓ Reliable: {len(reliable_df) / len(data_df) * 100:.2f}%")
print(f"  ✗ Unreliable: {(len(data_df) - len(reliable_df)) / len(data_df) * 100:.2f}%")

print("\n" + "=" * 60)
print("BREAKDOWN BY COLUMN")
print("=" * 60)

# Unreliable by type
unreliable_price = (data_df['reliable_price_and_unit_new'] == False).sum()
unreliable_promo = (data_df['reliable_promo_price_to_unit_price'] == False).sum()
unreliable_both = ((data_df['reliable_price_and_unit_new'] == False) & 
                   (data_df['reliable_promo_price_to_unit_price'] == False)).sum()

print(f"\nUnreliable due to price/unit_price: {unreliable_price:,} ({unreliable_price/len(data_df)*100:.2f}%)")
print(f"Unreliable due to promo_price: {unreliable_promo:,} ({unreliable_promo/len(data_df)*100:.2f}%)")
print(f"Unreliable due to BOTH: {unreliable_both:,} ({unreliable_both/len(data_df)*100:.2f}%)")

# Save cleaned DataFrame
print("\n" + "=" * 60)
reliable_df.to_csv(
    '../data/processed/data_df_reliable.csv',
    index=False
)
print("✓ Cleaned DataFrame saved as: data_df_reliable.csv")
print("=" * 60)
```
